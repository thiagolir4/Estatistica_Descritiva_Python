```python
#me devolve somente as colunas com o tipo numerico
# numerica continua = tem numeros decimais 
# numerica discreta = somente numeros inteiro
numericas = df.select_dtypes(include=['number'])
```

```python
#devolve somente variaveis categoricas
# categoricas nominal = somente para nomenclatura 
# categoricas ordinal = serve para algum tipo de classificação 
categoricas = df.select_dtypes(include=['object', 'category'])
```

```python
print(f"Variaveis categoricas {categoricas}")
```

```python
print(f"Variaveis numericas {numericas}")
```
