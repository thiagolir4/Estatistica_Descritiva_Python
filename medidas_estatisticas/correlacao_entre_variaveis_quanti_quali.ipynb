```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("./base_ifood_limpa.csv", sep=";", header=0)
```

```python
df['category'].value_counts()
```

```python
categorias_to_use = ['Lanches','Brasileira','Doces & Bolos','Pizza','Açaí']
df_tmp = df[df['category'].isin(categorias_to_use)]
```

```python
df_tmp.groupby("category")['minimumOrderValue'].agg(["mean","std"])
```

```python
df_tmp.groupby("category")['minimumOrderValue'].agg(["mean"]).plot.bar()
```

```python
df_tmp.boxplot(column="minimumOrderValue", by ="category")
```

```python
df.groupby("price_range")[["delivery_fee","delivery_time"]].agg(["mean"])
```

```python
df.groupby("price_range")[["delivery_fee"]].agg(["mean"]).plot.bar()
```

```python
df_tmp.boxplot(column="delivery_fee", by ="price_range")
```
