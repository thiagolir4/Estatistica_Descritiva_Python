```python
#vendo quantas linhas estão com NaN 
len(df[df["Cidade"].isna()])
```

```python
#sabendo a porcentagem que estão nulas na coluna cidade 
cidade_nula = len(df[df["Cidade"].isna()])/len(df) * 100
print(f"Percentual de Linhas Nulas na coluna Cidade: {cidade_nula:.2f}%")
```

```python
# fazendo com outras colunas 
list (df.columns)
```

```python
#desconbrindo a % de todas as colunas com linhas nulas 
colunas = list(df.columns)

for coluna in colunas:
    print(f"Percentual de linhas Nulas | Variável {coluna}: {100 * len(df[df[coluna].isna()])/len(df):.2f}%")
```
